```python
import numpy as np
```

```python
def correlation_matrix(X):
    """
    X: (N, d) numpy array
    """
    return np.corrcoef(X, rowvar=False)

# # Usage
# corr = correlation_matrix(X)

# # Optional: inspect magnitude
# print("Max |correlation off-diagonal|:",
#       np.max(np.abs(corr - np.eye(corr.shape[0]))))
```

```python
from sklearn.feature_selection import mutual_info_regression

def mutual_information_matrix(X):
    """
    Estimate pairwise mutual information matrix.
    """
    N, d = X.shape
    MI = np.zeros((d, d))
    
    for j in range(d):
        y = X[:, j]
        X_rest = X
        
        # mutual_info_regression computes I(y, X_k)
        mi = mutual_info_regression(X_rest, y, discrete_features=False)
        MI[j, :] = mi
    
    return MI

# Usage
# mi = mutual_information_matrix(X)

# print("Mean MI off-diagonal:",
#       np.mean(mi - np.diag(np.diag(mi))))
```

```python
def independence_summary(X):
    corr = correlation_matrix(X)
    mi = mutual_information_matrix(X)
    
    summary = {
        "max_abs_corr": np.max(np.abs(corr - np.eye(corr.shape[0]))),
        "mean_abs_corr": np.mean(np.abs(corr - np.eye(corr.shape[0]))),
        "mean_MI": np.mean(mi - np.diag(np.diag(mi))),
        "max_MI": np.max(mi - np.diag(np.diag(mi))),
    }
    
    return summary
```

```python
test_data = np.load("data_processed/test_data_scaled.npz")
X_test = test_data["x"]    

print(independence_summary(X_test))
```
