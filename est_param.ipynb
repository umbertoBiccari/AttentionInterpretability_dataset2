```python
import torch
import numpy as np
from math import prod
from models_transformer import SingleOutTransformerNet
```

```python
IN_DIM = 72
EMB_DIM = 64
NHEAD = 4
NUM_LAYERS = [3, 5, 7, 9, 11, 13, 15, 17, 20]
FF_DIM = 128
DROPOUT = 0.1
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

for layer in NUM_LAYERS:
    model = SingleOutTransformerNet(IN_DIM, emb_dim=EMB_DIM, nhead=NHEAD, 
                                    num_layers=layer, ff_dim=FF_DIM, 
                                    dropout=DROPOUT).to(DEVICE)

    params = [param.detach().numpy().shape for param in model.parameters()]
    n_params = sum([prod(p) for p in params])
    print(f"{layer} layers: {n_params} parameters")
```
