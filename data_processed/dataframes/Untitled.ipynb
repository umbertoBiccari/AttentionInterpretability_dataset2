```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("test_df_pruned.csv")
```

```python
bins = [11, 25, 40, 60, 80, 102]
labels = ["11-25", "25-40", "40-60", "60-80", "80-102"]
df["label_range"] = pd.cut(df["age"], bins=bins, labels=labels, include_lowest=True)
df_list = [df[df["label_range"] == lab] for lab in labels]
```
