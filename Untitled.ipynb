```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
from sklearn.metrics.pairwise import cosine_similarity
```

```python
results_rollout = np.load("Results/rollout_global/rollout.npz")
# results_rollout2 = np.load("Results/rollout_local/rollout.npz")
results_shap = np.load("Results/shap_values/shap_values.npz")
```

```python
mean_shap = np.mean(results_shap['shap_vals'], axis=0)
importance = results_rollout['rollout'].sum(axis=0)
```

```python
def l1_normalize(x):
    x = np.asarray(x, dtype=float)
    return x / np.sum(np.abs(x))

shap_norm = l1_normalize(mean_shap)
rollout_norm = l1_normalize(importance)

cos = cosine_similarity([shap_norm], [rollout_norm])[0, 0]
print("Cosine:", cos)
```
